```python
# round 2 처리
print((round(amazon['Sales Margin Amount'], 2) == round(amazon['Sales Amount'] - amazon['Sales Cost Amount'], 2)).value_counts())
print((round(amazon['Sales Price'] * amazon['Sales Quantity'], 2) == round(amazon['Sales Amount'], 2)).value_counts())
print((round(amazon['List Price'] * amazon['Sales Quantity'], 2) == round( amazon['Sales Amount Based on List Price'], 2)).value_counts())
```

```python
# 두 값의 차이 계산
difference = amazon['List Price'] * amazon['Sales Quantity'] - amazon['Sales Amount Based on List Price']

# 부동 소수점 오차로 간주할 수 있는 작은 차이를 확인 (예: 0.01보다 작은 경우)
floating_point_issues = difference.abs() < 0.01

print(floating_point_issues.value_counts())
```

```python
import numpy as np

# 부동 소수점 오차를 고려하여 비교
is_close = np.isclose(amazon['List Price'] * amazon['Sales Quantity'], amazon['Sales Amount Based on List Price'])

# True와 False 값의 개수 세기
true_cnt = np.sum(is_close)
false_cnt= len(is_close) - true_cnt

print(f"True: {true_cnt}, False:{false_cnt}")
```
