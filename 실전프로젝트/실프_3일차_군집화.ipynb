# box-cox

```python
import numpy as np
from scipy import stats

def boxcox_transform(column):
    transformed_column, best_lambda = stats.boxcox(column)
    return transformed_column, best_lambda

# 적용
for col in ['Recency', 'Frequency', 'Monetary']:
    transformed_col, best_lambda = boxcox_transform(cust_rfm[col])
    cust_rfm[f'{col}_box'] = transformed_col
    # print(f"{col}의 최적의 lambda 값: {best_lambda}")

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(12, 4), nrows=1, ncols=3)

# plt.ticklabel_format(style = 'plain')
ax1.set_title('Recency Histogram')
ax1.hist(cust_rfm['Recency_box'])

ax2.set_title('Frequency Histogram')
ax2.hist(cust_rfm['Frequency_box'])

ax3.set_title('Monetary Histogram')
ax3.hist(cust_rfm['Monetary_box'])
```

# Clustering

```python
cust_rfm.columns
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

features= ['Recency_box', 'Frequency_box', 'Monetary_box']

kmeans= KMeans(n_clusters= 5,init='k-means++', random_state= 42)
labels= kmeans.fit_predict(cust_rfm[features])
cust_rfm['cluster_label'] = labels

silhouette= silhouette_score(cust_rfm[features], labels)
print(f'실루엣 계수: {round(silhouette, 2)}')
```

실루엣 계수가 0.38 -> 개선해야 된다고 생각하기에는 아직 빠른 감이 있음.  
군집 간의 거리가 가까울 뿐 군집이 잘 돼있을 수도 있음!

각 군집간의 통계치, 각 데이터의 분포 등 -> 경향성, 패턴 찾기가 목표.

```python
silhouette_scores = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(cust_rfm[features])
    score = silhouette_score(cust_rfm[features], kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(range(2, 7), silhouette_scores, marker='o')
```

```python
def elbow(X):
    sse = [] # 군집 내 거리의 합
    for i in range(1, 11):
        km = KMeans(n_clusters=i, init='k-means++', random_state=42)
        km.fit(X)
        sse.append(km.inertia_)
    plt.plot(range(1, 11), sse, marker='o')
    plt.xlabel('클러스터 개수')
    plt.ylabel('SSE')
    plt.show()

elbow(cust_rfm[features])
```

```python
import copy

cust_rfm_copy= copy.deepcopy(cust_rfm)
```

```python
cust_rfm_copy.head(3)

cust_rfm_copy['cluster_label'].value_counts()
```

```python
from sklearn.mixture import GaussianMixture

# 데이터 준비
data = cust_rfm_copy[features]

# 클러스터 수 범위 설정
n_components_range = range(1, 11)  # 1부터 10까지의 클러스터 수
aic_values = []
bic_values = []

# GMM 모델 학습 및 AIC/BIC 계산
for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(data)
    aic_values.append(gmm.aic(data))
    bic_values.append(gmm.bic(data))

plt.figure(figsize=(10, 6))
plt.plot(n_components_range, aic_values, label='AIC', marker='o')
plt.plot(n_components_range, bic_values, label='BIC', marker='o')
plt.title('AIC and BIC for Different Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('AIC/BIC')
plt.xticks(n_components_range)
plt.legend()
plt.grid()
plt.show()

# 감소하다가 완만 -> 4?
```

```python
from sklearn.mixture import GaussianMixture

gmm= GaussianMixture(n_components= 4, random_state= 42)
gmm.fit(cust_rfm_copy[features])
gmm_cluster_label= gmm.predict(cust_rfm_copy[features])

cust_rfm_copy['gmm_cluster_label'] = gmm_cluster_label
```
