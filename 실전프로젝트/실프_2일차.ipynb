```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data

```python
amazon = pd.read_csv('/Users/t-2024-m0196/my-repo/Git/실전프로젝트/AmazonFoodCategory Dataset.csv')
print(amazon.info())
amazon.head(3)
```

```python
# 날짜 변수

amazon['DateKey'] = pd.to_datetime(amazon['DateKey'], format= '%m/%d/%Y')
amazon['Invoice Date'] = pd.to_datetime(amazon['Invoice Date'], format= '%Y/%m/%d')
amazon['Promised Delivery Date'] = pd.to_datetime(amazon['Promised Delivery Date'], format= '%m/%d/%Y')
# .sort_values(by= 'Datekey')
```

```python
amazon['DateKey'].sort_values()

# 2017 1 1 ~ 2019 12 31
```

```python
# 수치형 변수만 선택
numeric= amazon.select_dtypes(include= ['float64', 'int64'])

# 상관계수 행렬 계산
corr = numeric.corr()

# 히트맵
sns.heatmap(data= corr, annot= True, fmt= '.2f')
```
